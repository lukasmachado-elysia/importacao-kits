```python
import pandas as pd
import numpy as np
from funcoes_python.dba.banco import Banco
import datetime
import bisect
pd.set_option('display.float_format', lambda x: '%.3f' % x)
dateBackup = datetime.datetime.today().strftime("%Y_%m_%d")
```

```python
# Lendo arquivo com dados
df = pd.read_excel("C:/Users/Lukas/Downloads/Atualizacao_Kits (3).xlsx")
potenciaKvaTransformador = [int(i) for i in list('15,20,25,30,35,40,45,50,60,75,80,90,100,112,125,150,175,200,250,300'.split(','))]
precoTransformador = [float(i) for i in list('1998.77,2318.69,2911.77,3237.72,3583.28,3778.18,4366.96,4604.94,5460.68,6161.51,6528.64,7654.74,8477.15,8779.13,9316.73,10553.54,11694.15,12639.16,14811.40,16756.62'.split(','))]
```

```python
# Inserindo dados restantes
df['kwp_Sist'] = (df['quantidade_Modulos']*df['potencia_Modulos'])/1000
df['area'] = df['quantidade_Modulos']*2.56
df['eficiencia_Inversores'] = 98
df['potencia_Inversores'] = df['potencia_Inversores']*df['quantidade_Inversores']
df.loc[df['ligacao_Inversores'] == 'Monofásico', ['monitoramento']] = 'Fusion'
df.loc[df['ligacao_Inversores'] == 'Trifásico', ['monitoramento']] = 'SolarView'
df.loc[df['ligacao_Inversores'] == 'Monofásico', ['tensao_Inversores']] = int(220)
df.loc[df['ligacao_Inversores'] == 'Trifásico', ['tensao_Inversores']] = int(380)
df['marca_Estrutura'] = 'Sonnen'
df['reais_Watt'] = df['valor_Fibrocimento']/(df['kwp_Sist']*1000)
df['ton_co2'] = df['kwp_Sist']*0.5
df['arvores'] = df['kwp_Sist']*13 # arredondar
df['km_Evitados'] = df['kwp_Sist']*350 
df['garantia'] = 10 # checar
df['potencia_Aparente'] = df['potencia_Inversores']*1.1
```

```python
# Definindo os transformadores
potenciaAparente = list(df['potencia_Aparente'])
potenciaTransformadorDf = []
precoTransformadorDf = []
for pot in potenciaAparente:
    index = bisect.bisect(potenciaKvaTransformador,pot)
    # O indice nao pode ser maior que o tamanho da lista
    if index >= len(potenciaKvaTransformador):
        index = index -1
    potenciaTransformadorDf.append(int(potenciaKvaTransformador[index]))
    precoTransformadorDf.append(int(precoTransformador[index]))
```

```python
# Transformador e Precos
df['potencia_Transformador'] = potenciaTransformadorDf
df['preco_Transformador'] = precoTransformadorDf
df.loc[df['potencia_Aparente'] > df['potencia_Transformador'], ['potencia_Transformador','preco_Transformador']]  = ''
df.loc[df['ligacao_Inversores'] == 'Monofásico', ['potencia_Transformador','preco_Transformador']] = ''
```

```python
# Mes sul e Mes oeste
df['kwh_Sul'] = df['kWh_Mes(Geração_Norte)']/1.2962
df['kwh_Oeste'] = df['kWh_Mes(Geração_Norte)']/1.1290
df.round(decimals=1)
df = df.astype({'ton_co2':'int','km_Evitados':'int','arvores':'int'})
df
```

```python
# Reorganizando colunas
df = df[['kWh_Mes(Geração_Norte)',
        'kwp_Sist',
        'quantidade_Modulos',
        'marca_Modulos',
        'modelo_Modulos',
        'potencia_Modulos',
        'area',
        'quantidade_Inversores',
        'marca_Inversores',
        'modelo_Inversores',
        'potencia_Inversores',
        'eficiencia_Inversores',
        'monitoramento',
        'ligacao_Inversores',
        'tensao_Inversores', 
        'marca_Estrutura',
        'valor_Fibrocimento',
        'valor_Ceramico',
        'valor_Metalico',
        'valor_Laje',
        'valor_Solo',
        'reais_Watt',
        'ton_co2',
        'arvores',
        'km_Evitados',
        'garantia',
        'potencia_Aparente',
        'potencia_Transformador',
        'preco_Transformador',
        'kwh_Sul',
        'kwh_Oeste']]
```

```python
# Formatando comando sql
listString = []
listSQLCommand = []
for row in df.values:
    listString = []
    for column in row:
        # verificando se eh numero
        if isinstance(column, int) or isinstance(column, float):
            listString.append(str(column))
        else:
            listString.append('\'' + str(column) + '\'')
    listSQLCommand.append(','.join(listString))

listSQLCommand = ['(' + x + ')' for x in listSQLCommand]
insertSQL = ','.join(listSQLCommand)
```
